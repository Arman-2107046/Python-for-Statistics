HOUSING IN CALIFORNIA

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

Geographical Location

```python
df=pd.read_csv('housing.csv')
# df


g=sns.relplot(kind='scatter', data=df, x='longitude', y='latitude')

g.set(title='Geographical location of houses in California')

```

Median House Value

```python
g=sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', hue='median_house_value')

g.set(title='Geographical location of houses in California')
```

Ocean Proximity

```python
g=sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', hue='ocean_proximity')

g.set(title='Geographical location of houses in California')
```

```python
df['above_median_house_value']=(df.median_house_value> df.median_house_value.median()).astype(int)
# df
```

Above Median House

```python
g=sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', hue='above_median_house_value')

g.set(title='Geographical location of houses in California')
```

style

```python
g=sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', style='above_median_house_value',hue='above_median_house_value')

g.set(title='Geographical location of houses in California')
```

```python
g=sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', style='above_median_house_value',hue='above_median_house_value', size='median_house_value')

g.set(title='Geographical location of houses in California')
```
