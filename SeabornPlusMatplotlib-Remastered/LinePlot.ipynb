Line Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
sns.relplot(data=df, x='total_bill', y='tip',kind='line')
# sns.scatterplot(data=df, x='total_bill', y='tip')
```

```python
# sns.relplot(data=df, x='size', y='tip')
sns.relplot(data=df, x='size', y='tip', kind='line', hue='sex' ,col='sex')
```

```python
ndf=df.groupby('size')['tip'].sum()
print(ndf)
```

```python
sns.relplot(data=df,x='size', y='tip', kind='line',estimator='sum' ,ci=False, marker='o')
sns.relplot(data=df,x='size', y='tip', kind='line',estimator='max')
```
