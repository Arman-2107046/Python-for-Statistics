```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')

sns.catplot(data=df,
            x='day',
            y='tip')
sns.catplot(data=df,
            x='day',
            y='tip',
            kind='swarm',
            s=30,
            edgecolor='red',
            linewidth=.5,
            marker='o')

plt.show()
```
