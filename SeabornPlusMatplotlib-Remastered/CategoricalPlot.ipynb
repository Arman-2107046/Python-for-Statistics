CATEGORICAL PLOT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('taxis')
df.to_csv("taxis_csv")


sns.catplot(data=df, x='payment',y='total', jitter=0.3, height=5, aspect=2)
plt.show()
```

```python
sns.catplot(data=df, x='passengers', y='total',hue='payment', jitter=.4, height=10, aspect=2, dodge=True)
```

```python
sns.catplot(data=df, x='passengers', y='total',hue='payment', jitter=.4, height=10, aspect=2 , row='payment')
```

```python
sns.catplot(
    data=df,
    x='passengers',
    y='total',
    hue='color',
    palette={"yellow":'orange', "green":'blue'},
    jitter=.2,
    height=5,
    aspect=1.5,
    row='color',
    col='payment')
```

```python
sns.catplot(data=df, x='total')
plt.show()
```
