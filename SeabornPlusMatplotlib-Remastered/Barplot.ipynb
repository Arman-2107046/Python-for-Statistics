```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')

g=sns.catplot(
    data=df,
    x='size',
    y='tip',
    kind='bar',
    estimator='mean',
    errorbar=None,
    col='sex',
    hue='sex')

# xMean=df.groupby('size')['tip'].mean()
# print(xMean)

# xStd=df.groupby('size')['tip'].std()
# print(xStd)


for ax in g.axes.flat:
    for container in ax.containers:
        ax.bar_label(container)

plt.show()
```
