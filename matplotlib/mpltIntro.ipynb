BASIC_GRAPH

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Define the data points for the first plot
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize the graph
plt.figure(figsize=(5, 3), dpi=200)

# Plot the first line using shorthand notation
# 'b.-' means blue color, dot marker, and solid line
plt.plot(x, y, 'b.-', label='2x')

# Add title and labels for the axes
plt.title("Our First Graph!!", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.ylabel("Y-axis")
plt.xlabel("X-axis")

# Define the data points for the second plot
x2 = np.arange(0, 4, 0.5)

# Plot the second line up to the 6th value with a solid line
plt.plot(x2[:6], x2[:6]**2, 'r', label='y=x^2')

# Plot the second line from the 6th value onwards with a dashed line
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Set the ticks for the x and y axes
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

# Add a legend to the plot
plt.legend()

# Save the plot as an image file
plt.savefig('firstGraph.png', dpi=300)

# Display the plot
plt.show()
```

```python
labels=['A','B','C']
values=[1,4,2]

plt.figure(figsize=(6,4))

bars=plt.bar(labels,values)

patterns=['/','O','*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

plt.show()
```
